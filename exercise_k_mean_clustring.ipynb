```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
iris = datasets.load_iris()
x = iris.data[:,2:4]
```

```python
# creat list of k * (1-10)
# creat empty list to store score , name it score_lst
# loop on k:
# creat kmeans model
# fit model
# get score * (silhouette score) and append to score_lst
# plot scatter k and score

#data = range(list(1, 11))
score_lst = []

for i in range(2,11):
  kmeans = KMeans(n_clusters=i, random_state=42)
  kmeans.fit(x)
  labels = kmeans.labels_
  centers = kmeans.cluster_centers_

  score = silhouette_score(x,labels)
  score_lst.append(score)
print('silhouutte score: ', score_lst)


plt.scatter(x[:,0],x[:,1], c=labels, cmap = 'viridis')
plt.scatter(centers[:,0],centers[:,1] , c='red', s=200 , alpha=0.5)
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.title('k-mean clustring with k=3')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression


```

```python
#loud dataset
data = fetch_california_housing()
x = data.data
y = data.target
```

```python
# split data into traing / testing sets
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# train / validation split

model = LinearRegression()
model.fit(x_train, y_train)

# vlidate the model
score_train_val = model.score(x_test, y_test)
print("Train / validation score :  ",{score_train_val})
```

```python
# k-fold cross validation
scores_cv = cross_val_score(LinearRegression(), x,y,cv=10)

# print each of the score and average score
print(f"score validation scores: {scores_cv}")
print(f"avarege cross validation scores: {np.mean(scores_cv)}")
```

```python
#vitulize scores
plt.figure(figsize=(12,6))
plt.bar(range(1,11),scores_cv,color='Gray', label='cross validation score')
plt.axhline(y=score_train_val, color="red", linestyle='--',label='Train Validation score')
plt.axhline(y=np.mean(scores_cv), color="pink", linestyle='--',label='Train Validation score')
plt.xlabel=('Flode Number')
plt.ylabel=('Score')
plt.legend()
plt.title('train / validation')
plt.show()


```
